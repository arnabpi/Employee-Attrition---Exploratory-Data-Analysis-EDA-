### Summary of Data


There are 1470 rows and 20 Columns in the dataset.

- Columns & their Meaning
    1. Age(Numerical) - Age's of Employees in this company
    2. Attrition - Weather the employee has left the company or not
    3. Business Travel(Categorical/Ordinal) - Travel Rate of the Employee in the company
    4. Department(Categorical) - Different departments in the company
    5. Education Field(Categorical) - Field of study for each employee in the company
    6. EnvironmentSatisfaction(Numerical/Ordinal) - Are they satisfied by the company's evirnoment
    7. Gender(Categorical) - Gender of Employee in the company
    8. Job Involvement(Numerical/Ordinal) - 
    9. Job Level(Numerical/Ordinal) - Position level of employee in the company
    10. Job Statisfaction(Numerical/Ordinal) - Are the employee satisfied by the job they are doing in company
    11. Maritial Status(Categorical) - Weather the employee are married or not in the company
    12. Monthly Income(Numerical/Ratio) - Income of each employee in the company
    13. Overtime(Categorical) - Overtime culture in the company
    14. Total Working Years(Numerical/Ordinal) - How many years of experience does the employee have
    15. Work Life Balance(Numerical/Ordinal) - Is there any work life balance in the company
    16. Years at Company(Numerical) - How many years does the employee have worked in the company
    17. Years in Current Role(Numerical) - What is the current role on employee in the company
    18. Years Since last promotion(Numerical) - When was the employee was last promoted
    19. Years with current Manager(Numerical) - Years spent by each employee with current manager in the company
    20. Distance From Home(Numerical) - Distance of cummute for each employee in the company
    
    
Most features in this dataset are categorical as well as ordinal i.e orders matter. The lower the ordinal variable the worse it is in that case.
Ex: Job Satisfaction 1 = "Low" and 4 = "High"
